```python
import tweepy
from tweepy import OAuthHandler, Stream, StreamListener

consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
tweets = api.home_timeline()
for tweet in tweets:
    print(tweet.text)
```

```python
user_tweets = api.user_timeline(screen_name='jokowi', count=200, tweet_mode="extended")
for tweet in user_tweets:
    print(tweet._json)
```

```python
import datetime
import pandas
user_tweets_list = []

user_tweets = api.user_timeline(screen_name='AdjieSanPutro', count=200, tweet_mode="extended")
startDate = datetime.datetime(2019, 10, 22, 0, 0, 0)
endDate = datetime.datetime(2019, 10, 24, 0, 0, 0)
for tweet in user_tweets:
    if tweet.created_at < endDate and tweet.created_at > startDate:
        user_tweets_list.append(tweet._json)
        
userData = pandas.DataFrame(user_tweets_list)
userData
```

```python
hashtags = ["#google", "#100DaysOfCode", "#python"]
hashtag_tweets = []

for hashtag in hashtags:
    for tweet in api.search(q=hashtag, lang="en", rpp=50, tweet_mode="extended"):
        hashtag_tweets.append(tweet._json)

hashtagData = pandas.DataFrame(hashtag_tweets)
hashtagData.to_csv("hashtag.csv")
hashtagData
```
